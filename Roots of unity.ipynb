```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, IntSlider

def plot_roots_of_unity(n):
    # Calculate the n-th roots of unity
    roots = [np.exp(2j * np.pi * k / n) for k in range(n)]

    # Extract the real and imaginary parts
    reals = [root.real for root in roots]
    imags = [root.imag for root in roots]

    # Plot the unit circle
    circle = np.exp(2j * np.pi * np.linspace(0, 1, 1000))
    plt.plot(circle.real, circle.imag, 'b--', label='Unit Circle')

    # Plot the roots of unity
    plt.scatter(reals, imags, color='red', marker='o', s=100, label=f'{n}th Roots of Unity')
    
    # Connect each root to the next one with a line
    for i in range(n):
        plt.plot([reals[i], reals[(i+1) % n]], [imags[i], imags[(i+1) % n]], 'g-')
        plt.text(reals[i], imags[i], f'({reals[i]:.2f}, {imags[i]:.2f})', fontsize=9, ha='right')

    # Set plot details
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.title(f'{n}th Roots of Unity')
    plt.xlabel('Real Part')
    plt.ylabel('Imaginary Part')
    plt.legend()
    plt.grid(True)
    plt.show()

# Create an interactive slider to control the value of n
interact(plot_roots_of_unity, n=IntSlider(min=1, max=30, step=1, value=2))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, IntSlider

def plot_roots_of_unity(n):
    # Calculate the n-th roots of unity
    roots = [np.exp(2j * np.pi * k / n) for k in range(n)]

    # Extract the real and imaginary parts
    reals = [root.real for root in roots]
    imags = [root.imag for root in roots]

    # Plot the unit circle
    circle = np.exp(2j * np.pi * np.linspace(0, 1, 1000))
    plt.plot(circle.real, circle.imag, 'b--', label='Unit Circle')

    # Plot the roots of unity
    plt.scatter(reals, imags, color='red', marker='o', s=100, label=f'{n}th Roots of Unity')

    # Connect each root to the next one with a line
    side_lengths = []
    for i in range(n):
        next_index = (i + 1) % n
        plt.plot([reals[i], reals[next_index]], [imags[i], imags[next_index]], 'g-')
        plt.text(reals[i], imags[i], f'({reals[i]:.2f}, {imags[i]:.2f})', fontsize=9, ha='right')
        # Calculate the side length
        side_length = np.sqrt((reals[i] - reals[next_index])**2 + (imags[i] - imags[next_index])**2)
        side_lengths.append(side_length)

    # Estimate π
    perimeter = sum(side_lengths)
    pi_estimate = perimeter / 2

    # Set plot details
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.title(f'{n}th Roots of Unity\nEstimated π: {pi_estimate:.6f}')
    plt.xlabel('Real Part')
    plt.ylabel('Imaginary Part')
    plt.legend()
    plt.grid(True)
    plt.show()

# Create an interactive slider to control the value of n
interact(plot_roots_of_unity, n=IntSlider(min=1, max=50, step=1, value=2))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import FloatSlider, interact

# Define the plotting function
def plot_euler_formula(t):
    x = np.linspace(0, 2 * np.pi, 400)
    real_part = np.cos(t * x)
    imaginary_part = np.sin(t * x)
    
    plt.figure(figsize=(12, 12))  # Adjust the figsize parameters here
    plt.plot(real_part, imaginary_part, label=f"Unit Circle (t={t})", color="purple")
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(color='gray', linestyle='--', linewidth=0.5)
    plt.title("Unit Circle Representation of $e^{i(t \cdot x)}$")
    plt.xlabel("Real Part")
    plt.ylabel("Imaginary Part")
    plt.legend()
    plt.gca().set_aspect('equal', adjustable='box')
    
    # Fix the x and y axis limits
    plt.xlim(-1.5, 1.5)
    plt.ylim(-1.5, 1.5)
    
    plt.show()

# Create the FloatSlider
slider = FloatSlider(value=0.5, min=0.0, max=1.0, step=0.01, description='t', continuous_update=True)

# Use interact to link the slider with the plotting function
interact(plot_euler_formula, t=slider)
```
