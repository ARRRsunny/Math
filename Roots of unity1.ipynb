```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, IntSlider

def plot_roots_of_unity(n):
    # Calculate the n-th roots of unity
    roots = [np.exp(2j * np.pi * k / n) for k in range(n)]

    # Extract the real and imaginary parts
    reals = [root.real for root in roots]
    imags = [root.imag for root in roots]

    # Plot the unit circle
    circle = np.exp(2j * np.pi * np.linspace(0, 1, 1000))
    plt.plot(circle.real, circle.imag, 'b--', label='Unit Circle')

    # Plot the roots of unity
    plt.scatter(reals, imags, color='red', marker='o', s=100, label=f'{n}th Roots of Unity')
    
    # Connect each root to the next one with a line
    for i in range(n):
        plt.plot([reals[i], reals[(i+1) % n]], [imags[i], imags[(i+1) % n]], 'g-')
        plt.text(reals[i], imags[i], f'({reals[i]:.2f}, {imags[i]:.2f})', fontsize=9, ha='right')

    # Set plot details
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.title(f'{n}th Roots of Unity')
    plt.xlabel('Real Part')
    plt.ylabel('Imaginary Part')
    plt.legend()
    plt.grid(True)
    plt.show()

# Create an interactive slider to control the value of n
interact(plot_roots_of_unity, n=IntSlider(min=1, max=30, step=1, value=2))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, IntSlider

def plot_roots_of_unity(n):
    # Calculate the n-th roots of unity
    roots = [np.exp(2j * np.pi * k / n) for k in range(n)]

    # Extract the real and imaginary parts
    reals = [root.real for root in roots]
    imags = [root.imag for root in roots]

    # Plot the unit circle
    circle = np.exp(2j * np.pi * np.linspace(0, 1, 1000))
    plt.plot(circle.real, circle.imag, 'b--', label='Unit Circle')

    # Plot the roots of unity
    plt.scatter(reals, imags, color='red', marker='o', s=100, label=f'{n}th Roots of Unity')

    # Connect each root to the next one with a line
    side_lengths = []
    for i in range(n):
        next_index = (i + 1) % n
        plt.plot([reals[i], reals[next_index]], [imags[i], imags[next_index]], 'g-')
        plt.text(reals[i], imags[i], f'({reals[i]:.2f}, {imags[i]:.2f})', fontsize=9, ha='right')
        # Calculate the side length
        side_length = np.sqrt((reals[i] - reals[next_index])**2 + (imags[i] - imags[next_index])**2)
        side_lengths.append(side_length)

    # Estimate π
    perimeter = sum(side_lengths)
    pi_estimate = perimeter / 2

    # Set plot details
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.title(f'{n}th Roots of Unity\nEstimated π: {pi_estimate:.6f}')
    plt.xlabel('Real Part')
    plt.ylabel('Imaginary Part')
    plt.legend()
    plt.grid(True)
    plt.show()

# Create an interactive slider to control the value of n
interact(plot_roots_of_unity, n=IntSlider(min=1, max=50, step=1, value=2))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of x values
x = np.linspace(-10, 10, 400)

# Calculate the real and imaginary parts
real_part = np.cos(x)
imaginary_part = np.sin(x)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x, real_part, label="Real Part: cos(x)", color="blue")
plt.plot(x, imaginary_part, label="Imaginary Part: sin(x)", color="orange")
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.title("Euler's Formula: $e^{ix} = \cos(x) + i\sin(x)$")
plt.xlabel("x")
plt.ylabel("Value")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import FloatSlider, interact

# Define the plotting function
def plot_euler_formula(t):
    x = np.linspace(0, 2 * np.pi, 400)
    real_part = np.cos(t * x)
    imaginary_part = np.sin(t * x)
    
    plt.figure(figsize=(12, 12))  # Adjust the figsize parameters here
    plt.plot(real_part, imaginary_part, label=f"Unit Circle (t={t})", color="purple")
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(color='gray', linestyle='--', linewidth=0.5)
    plt.title("Unit Circle Representation of $e^{i(t \cdot x)}$")
    plt.xlabel("Real Part")
    plt.ylabel("Imaginary Part")
    plt.legend()
    plt.gca().set_aspect('equal', adjustable='box')
    
    # Fix the x and y axis limits
    plt.xlim(-1.5, 1.5)
    plt.ylim(-1.5, 1.5)
    
    plt.show()

# Create the FloatSlider
slider = FloatSlider(value=0.5, min=0.0, max=1.0, step=0.01, description='t', continuous_update=True)

# Use interact to link the slider with the plotting function
interact(plot_euler_formula, t=slider)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact
import math  # Import the standard library math module

# Function to calculate the partial sums
def taylor_series_approx(x, B):
    return sum([x**n / math.factorial(n) for n in range(B + 1)])

# Interactive plot function
def plot_taylor_series(B):
    x = np.linspace(-5, 5, 400)
    y_exp = np.exp(x)
    y_partial_sum = np.array([taylor_series_approx(val, B) for val in x])
    
    plt.figure(figsize=(10, 6))
    plt.plot(x, y_exp, label='$e^x$', color='black', linewidth=2)
    plt.plot(x, y_partial_sum, label=f'$\\sum_{{n=0}}^{{{B}}} \\frac{{x^n}}{{n!}}$', color='blue')
    plt.title('Taylor Series Approximation of $e^x$')
    plt.xlabel('$x$')
    plt.ylabel('$y$')
    plt.legend()
    plt.grid(True)

    plt.xlim(-5,5)
    plt.ylim(-5,5)
    plt.show()

# Create the interactive slider
interact(plot_taylor_series, B=(0, 20, 1))
```

```python
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact
import math

# Function to calculate the Taylor series approximation for ln(1+x)
def taylor_ln_approx(x, N):
    terms = [(-1)**(n+1) * x**n / n for n in range(1, N + 1)]
    return sum(terms)

# Interactive plot function
def plot_taylor_ln(N):
    x = np.linspace(-0.99, 3, 400)  # The series converges for |x| < 1
    y_ln = np.log(1 + x)
    y_taylor = np.array([taylor_ln_approx(val, N) for val in x])
    
    plt.figure(figsize=(10, 6))
    plt.plot(x, y_ln, label='$\ln(1+x)$', color='black', linewidth=2)
    plt.plot(x, y_taylor, label=f'$\\sum_{{n=1}}^{{{N}}} \\frac{{(-1)^{{n+1}} x^n}}{{n}}$')
    plt.title('Taylor Series Approximation of $\ln(1+x)$')
    plt.xlabel('$x$')
    plt.ylabel('$y$')
    plt.legend()
    plt.grid(True)
    plt.xlim(-3,3)
    plt.ylim(-3,3)
    plt.show()

# Create the interactive slider
interact(plot_taylor_ln, N=(1, 40, 1))
```

```python
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact
import math

# Function to calculate the Taylor series approximation for sin(x)
def taylor_sin_approx(x, N):
    terms = [(-1)**n * x**(2*n+1) / math.factorial(2*n+1) for n in range(N + 1)]
    return sum(terms)

# Interactive plot function
def plot_taylor_sin(N):
    x = np.linspace(-10*np.pi, 10*np.pi, 400)
    y_sin = np.sin(x)
    y_taylor = np.array([taylor_sin_approx(val, N) for val in x])
    
    plt.figure(figsize=(10, 6))
    plt.plot(x, y_sin, label='$\sin(x)$', color='black', linewidth=2)
    plt.plot(x, y_taylor, label=f'Taylor Series (N={N})', linestyle='--')
    
    plt.title('Taylor Series Approximation of $\sin(x)$')
    plt.xlabel('$x$')
    plt.ylabel('$y$')
    plt.legend()
    plt.xlim(-10,10)
    plt.ylim(-3,3) 
    plt.grid(True)
    plt.show()

# Create the interactive slider
interact(plot_taylor_sin, N=(0, 13, 1))
```
